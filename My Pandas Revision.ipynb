# Filtered Index

```python
filt = ['Country', 'Student', 'EdLevel', 'EduOther']
```

```python
# The 'loc' method was used to get the range of values by 0 through 10 
# Using the 'filt' variable has a filter for the columns
df.loc[0:10, filt]
```

```python
# Lets take it a level higher
# Get the students with a response of 'Yes, full-time' in the same list and not having values of NaN in 'EduOther'
filt2 = (df['Student'] == 'Yes, full-time')
```

```python
# Calling 'filt2' will give values in boolean
# Using the 'loc' will get you the values
# Check the data frames and confirm that the values of 'Student' must be 'Yes, full-time'
df.loc[filt2].head(5)
```

```python
# Now lets combine it with our 'filt' variable
# Now the condition is been added to the data frame
df.loc[filt2, filt].head(5)
```

```python
# Trying to get the Country, the language worked with be 'Python', Their years of code and type of developer they are
# Using the 'str', 'contains' and 'isin' func we will confirm the language contains "Python"
# 'contains' work with 'str' like: 'str.contains()'
# 'na = False' is to get values that are not 'NaN'
filt3 = df['LanguageWorkedWith'].str.contains("Python", na = False)
```

```python
filt3
```

```python
# Finally we've gotten what we aimed for!!!
prefered_data = ['Country', 'LanguageWorkedWith', 'YearsCode', 'DevType']
df.loc[filt3, prefered_data].head(10)
```
