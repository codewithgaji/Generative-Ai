# Grouping the data

First - SPlitting our objects

```python
df['Country'].value_counts()
```

```python
# Let's first group all of the results by 'Country' before splitting
# We assign the func to a variable called 'country_grp' so we don't have to 
# rewrite the code everytime
country_grp = df.groupby(['Country'])
```

```python
# '.get_group' is a func for 'groupby'

# Now we can get a specific group by country name
# Now check the data frame below and you will realise that they all are from..
# United States
country_grp.get_group('United States').head(5)# Getting the group for 'United States'
```

```python
country_grp.get_group('India').head(5)# Getting the group for 'India'
```

```python
# We can as well use the 'FILTER' method that we've learnt in previous lessons
filt = df['Country'] == 'China'
df.loc[filt].head(3)
```

WHY IS 'groupby' BETTER THAN FILTER METHOD?

```python
# IF YOU ARE PLANNING ON GETTING ONLY ONE COUNTRY IN THE ROWS....
# FILTERING IS BETTER.
# IF YOU WANT AN ALREADY BROKEN DOWN/SPLITTED DATA WHERE IT IS GROUPED BY EACH....
# ...COUNTRY, USING THE GROUP FUNCTION WOULD BE YOUR BEST OPTION FOR EASIER..
# MANIPULATION
```

APPLYING FUNCTIONS TO THE SPLITTED DATA 

```python
# Here we can now get the highest social media used for 'United States'
# because of the filter variable
filt = df['Country'] == 'United States'
df.loc[filt]['SocialMedia'].value_counts().head(5)
```

IF YOU NOTICED, FILTER(here) WORKS FOR ONE SPECIFIC COUNTRY LIKE(UNITED STATES)

NOW LET'S USE THE 'groupby' METHOD AND GET THE HIGHEST USED SOCIAL MEDIA FOR EVERY COUNTRY

```python
# Don't forget the 'country_grp' variable we created for the 'groupby' method above
# Now we can see all the social media sites grouped by their countries
country_grp['SocialMedia'].value_counts().head(50)
```

```python
# We can even specify the series that we want
# You might be wondering how is this useful if I can get the same results with filter?
# Well, It's useful because now you can easily search for any country....
# ...and Check their most used 'SocialMedia' by using '.loc['']'
# ...because 'groupby' function readily seperates the rows for you.
country_grp['SocialMedia'].value_counts().loc['India'] # specifying for india here
```
